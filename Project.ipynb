3. Для удобства работы с жанрами заменим тип параметра artist_genre со строки на список строк.

```python
def convert_genre(genres):
  genres = genres.split(',')
  for g in range(len(genres)):
    genres[g] = genres[g].replace('[', '')
    genres[g] = genres[g].replace(']', '')
    genres[g] = genres[g].replace("'", '')
    genres[g] = genres[g].strip()
  genres.sort()
  return genres

song_data['artist_genre'] =[convert_genre(i) for i in song_data['artist_genre']]
```

```python
genres = pd.DataFrame(columns=['genre'])
for i in range(500):
  for g in song_data['artist_genre'].iloc[i]:
    genres.loc[len(genres.index)] = [g]

genres = genres.agg(['value_counts'])['genre']
smaller_genres = pd.DataFrame(columns=['genre', 'count'])
for i in genres.index:
  smaller_genres.loc[len(smaller_genres.index)] = [
      'other' if genres['value_counts'].loc[i] < 20 else i,
        genres['value_counts'].loc[i]
  ]

fig = px.pie(smaller_genres,
             values=smaller_genres['count'],
             names=smaller_genres['genre'],
             color_discrete_sequence=px.colors.sequential.Agsunset,
             title="Распределение жанров")

fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

Тут очень много разных поджанров (большая часть из них попала в раздел другое), давайте выделем их основы и посмотрим тенденции в основах.

```python
def get_pure(g):
  pure_genres = ['pop', 'rap', 'hip hop', 'edm', 'reggaeton', 'r&b']
  f = None
  for i in pure_genres:
    if i in g:
      f = i
      break
  return i if f != None else 'other'

pure_genre_df = pd.DataFrame(columns=['genre', 'count'])
for i in genres.index:
  pg = get_pure(i)
  if i in pure_genre_df['genre']:
      pure_genre_df[pure_genre_df['genre'] == pg]['count'] += genres['value_counts'].loc[i]
  else:
    pure_genre_df.loc[len(pure_genre_df.index)] = [
        pg,
        genres['value_counts'].loc[i]
    ]
    

fig = px.pie(pure_genre_df,
             values=pure_genre_df['count'],
             names=pure_genre_df['genre'],
             color_discrete_sequence=px.colors.sequential.Agsunset,
             title="Распределение жанров")
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

Выделали 6 самых поппулярных основ - 'pop', 'rap', 'hip hop', 'edm', 'reggaeton', 'r&b'. \
Самый часто встречающийся жанр - pop (почти половина от всех), далее rap, hip hop, и остальные. \
Pop (от английского popular) простой и понятный многим жанр, не удивительно что он занимает половину всех жанров исполнителей.

Собираем данные об авторах статей с хабра https://habr.com/ru/users/

```python
import requests
import datetime
from bs4 import BeautifulSoup as bs
```

Собираем ссылки на страницы авторов с первых 5 страниц списка авторов.

```python
links = []
for i in range(1, 6):
  request = requests.get(f'https://habr.com/ru/users/page{i}/')
  if not request.ok:
    print('Error')
  html = bs(request.text, 'html.parser')
  users = html.findAll('div', {'class':'tm-users-list__item'})
  for user in users:
    links.append(user.findAll('a')[0]['href'])
```

```python
print(len(links))
```

Получили ровно 100 авторов

```python
# Небольшая функция вычленения информации со страницы автора.
def GetUser(url0):
  page0 = requests.get(url0)
  soup0 = bs(page0.text, 'html.parser')

  username = soup0.findAll("a", class_='tm-user-card__nickname')[0].text.strip() if len(soup0.findAll("a", class_='tm-user-card__nickname')) > 0 else None
  karma = soup0.findAll("div", class_='tm-karma__votes')[0].text.strip() if len(soup0.findAll("div", class_='tm-karma__votes')) > 0 else None
  about = soup0.findAll("p", class_='tm-user-card__short-info')[0].text.strip()if len(soup0.findAll("p", class_='tm-user-card__short-info')) > 0 else None
  rating = soup0.findAll("div", class_='tm-rating__counter')[0].text.strip() if len(soup0.findAll("div", class_='tm-rating__counter')) > 0 else None
  
  position_in_rating = soup0.findAll("dd", class_='tm-description-list__body tm-description-list__body tm-description-list__body_variant-base')[0].text.strip() if len(soup0.findAll("dd", class_='tm-description-list__body tm-description-list__body tm-description-list__body_variant-base')) > 0 else None
  if position_in_rating != None:
    position_in_rating = int(position_in_rating[:-2])
  
  # Вычленяем дату
  registration_date = datetime.datetime.strptime(soup0.findAll("time")[0]['title'].split(',')[0], '%Y-%m-%d') if len(soup0.findAll("time")) > 0 else None

  publishings = soup0.findAll("span", class_='tm-tabs__tab-counter')[0].text.strip() if len(soup0.findAll("span", class_='tm-tabs__tab-counter')) > 0 else None
  if publishings != None:
    # если у пользователя более тысячи публикаций, общее количество указывается в тысячах с символом К
    if 'K'in publishings:
      publishings = int(float(publishings[:-1])*1000)
    else:
      publishings = int(publishings)

  comments = soup0.findAll("span", class_='tm-tabs__tab-counter')[1].text.strip() if len(soup0.findAll("span", class_='tm-tabs__tab-counter')) > 1 else None
  if comments != None:
    # если у пользователя более тысячи коментариев, общее количество указывается в тысячах с символом К
    if 'K'in comments:
      comments = int(float(comments[:-1])*1000)
    else:
      comments = int(comments)

  badges = []
  # собираем значки
  for badge in soup0.findAll("li", class_='tm-badges__item'):
    badges.append(badge.text.strip())
    
  # собираем хабы
  hubs = []
  for hub in soup0.findAll("a", class_='tm-user-hubs__hub'):
    hubs.append('https://habr.com'+hub['href'])
  return username, about, float(karma), float(rating), position_in_rating, registration_date, publishings, comments, ', '.join(badges), ', '.join(hubs)
```
